```python
# gemini model as regular LLM

rmodel = GoogleGenerativeAI(model="gemini-2.5-flash")

# GoogleGenerativeAI parameters: model, api_key, max_tokens, temperature, retries,request_timeout
```

```python
#GoogleGenerativeAI(Parameter): max_tokens:

#what is token: It is basic unit which your LLM process and generate one by one.
#tokens: small pieces of the big things. Tokens can be a[word,character, sub-word]
#Character level tokenization: e.g gemini ==["g","e","m","i","n","i"]
#Word Level tokenization: e.g " Gemini is good model" == ["Gemini","is","good","model"].

#max_tokens: Use of max_tokens is to restrict the usage of tokens.
#max_tokens:10 .. maximum tokens we are able to use here are 10.
#Here gemini can't support max_token, gemini supports another parameter called as max_output_tokens

# Temperature: It is a technique to bring randomness to the LLM output.
# this temperature will take values from zero to one.
# when we use temperature is zero: then we telling the LMM that LMM you are not a random model, you are Deterministic model (means:for same query you are going to generate always one same answer).
# when we use temperature is one: To make your LLM more creative with same question(means: same question different different outputs)
# Always go with 0.5 not 0 and 1.
```

```python
prompt="tell me about agentic AI in one sentence"
```

```python
# call the model-- invocation : it is the technique to call the LLM to generate the output

# It has 3 types : invoke(), stream(), batch()

#invoke(): when ever we want to send a single query(prompt) : give me output after preforming the concatenation.

#stream(): when ever we want to send a single query(prompt) : the output token by token (ChatGPT,Google Gemini, Perplexity use this concept:stream)
#- the output token by token as it will give me an iterator object

#batch() #whenever we want to ask multiple query at a time.
```

```python
#here use retries,request_timeout parameters.
rmodel = GoogleGenerativeAI(model="gemini-2.5-flash",temperature=0,retries=3,request_timeout=10)

# retries :This tells the system how many times to try again if the request fails due to network or temporary issues.

# Example: retries=3 → If API fails, it will try up to 3 more times.

# request_timeout :This controls how long your code should wait for the API to respond before giving up.

# Example: request_timeout=30 → wait 30 seconds max.

#Think of it like: If the model takes too long, your program says "time’s up!" and stops waiting.
```

```python
#using stream()
resultstr = rmodel.stream("who is ms dhoni")
```

```python
resultstr
```

```python
for tokens in rmodel.stream("who is ms dhoni"):
  print(tokens)
```

```python
#using Batch()

# Used to send many prompts together in one job.

# Response comes after processing all items (not instant).

# Best for bulk tasks like summarizing 100 documents or analyzing lots of data.
```

```python
answer = rmodel.batch(["what is the full form of ai","what is the full form of DL"])
```

```python
answer=[]
for prompt in ["what is the full form of ai","what is the full form of DL"]:
  answer.append(rmodel.invoke(prompt))
```

```python
rmodel.map().invoke(["what is the full form of ai","what is the full form of DL"])
```

```python
#Invoke is best when you want a quick, real-time response for a single input, like in chatbots or automation steps. Batch is best when you need to process many inputs together, like summarizing hundreds of documents or analyzing a large dataset. So both are useful — Invoke for speed, Batch for bulk processing and efficiency.
```
